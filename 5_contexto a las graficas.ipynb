Agreamos títulos, etiquetas, tamaños y leyendas para dar contexto y mejor información a las graficas 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(8,5))
axes[0].plot(x,y, label='$sin(x)$') # $ --> agrego la etiquta a modo de notación matematica
axes[0].set_title('Relación X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x, 'r') 
axes[1].set_title('Relación Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')

#fig.tight_layout()
fig.show()
```

Sin utilizar el metodo orientado a objetos

```python
plt.plot(x,y, label='sin(x)')
plt.title('Titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
